dataframe structure changes

```python
#rename columns
df.rename(columns={"type1":"typex"}).head() #, inplace=True)
```

```python
#add new column
df["att_def"] = df["attack"]+df["defense"]
df.columns
```

```python
#drop columns
df.drop(columns = ["att_def"], inplace= True)
```

```python
#sort values
df.sort_values(by=['attack', 'defense'], ascending = False) #inplace=True
```

```python
#filter data
df[(df["type1"]== 'Dragon') & (df["legendary"] == True) & (df.attack > 150)]
# make sure to use parenthesis else we get an error esp with bool
```

```python
#create new dataframes
df1 = df[df["type2"] == 'Flying']
df2 = df[df["type2"] == 'Ice']
print("records in df1 = ", len(df1))
print("records in df2 = ", len(df2))
```

```python
#merge dataframes
df_merged = df1.merge(df2, on='number', how= 'left')
#df_merged = df1.join(df2, on='number', how= 'left') #has restrictions on the column it is joined on, might have to rename
```

```python
#delete dataframes
del df1, df2
```

missing values

```python
#identify missing values
#df.isnull().sum()
df.type2.isnull().sum()
```

```python
#fill missing values - insert values/mean,median,mode/interpolate (time series)
df['type2'].fillna(df['total'].mode()[0]) #.isnull().sum() #for median use .median() and for mean uise .mean()
#df['type2'].fillna(method = 'ffill') #or bfill
```

data aggregation

```python
#group by and aggregation
df.groupby('legendary').agg(sum_total = ("total", "sum"), countd_attack = ("attack", lambda x:x.nunique()), count_num = ("number", "count"))
```

```python
#create a binned value set for the defense column
min_val = df["defense"].min()
max_val = df['defense'].max()

bin_width = (max_val - min_val)/4 #calculate the width of each bin: want 4 equal parts
print("bin width = ", bin_width)

bins = [min_val + i * bin_width for i in range(1,5)]
print("bins = ", bins)

bins.insert(0, min_val) # insert min_val as the first value here
print("bins = ", bins)

#bins = [round(x) for x in bins]
#print("rounded bins = ", bins)

df["binned_def"] = pd.cut(df["defense"], bins, labels=["low", "medium", "high", "very high"])

df.head()
```

```python
#pivot table
df.pivot_table(values= "number", index= ["type2", "legendary"], columns = "binned_def", aggfunc="count")
```
